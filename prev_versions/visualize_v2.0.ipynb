## 2. Load History Files

### Supporting Functions

```python
class dotdict(dict):    
    """dot.notation access to dictionary attributes"""      
    def __getattr__(*args):        
        val = dict.get(*args)         
        return dotdict(val) if type(val) is dict else val     
    __setattr__ = dict.__setitem__     
    __delattr__ = dict.__delitem__

def alpha_sort(iterable): 
    """ Sort the given iterable in the way that humans expect.""" 
    convert = lambda text: int(text) if text.isdigit() else text 
    alphanum_key = lambda key: [ convert(c) for c in re.split('([0-9]+)', key) ] 
    return sorted(iterable, key = alphanum_key)

class_labels = {'dbpedia_14':{0:'Company', 1:'Edu Institution', 2:'Artist', 3:'Athlete', 4:'Office Holder',5:'Mean Of Trans', 6:'Building',\
                              7:'Natural Place', 8:'Village', 9:'Animal', 10:'Plant', 11:'Album', 12:'Film', 13:'Written Work'},
                'yahoo_answers':{0:"Society", 1:"Science", 2:"Health", 3:"Education", 4:"Computers",\
                                    5:"Sports", 6:"Business", 7:"Music", 8:"Family", 9:"Politics"}}
```

### Consolidate

```python
history_dir = './history'
history_dict = dict()
for path, dirs, files in os.walk(history_dir):
    if not dirs:
        dataset = path.split('/')[-2]
        strategy = path.split('/')[-1]
        for file_ in files:
            if dataset in history_dict:
                history_dict[dataset][strategy] = []
            else:
                history_dict[dataset] = {strategy:[]}
                
            for file_ in alpha_sort(files):
                history_dict[dataset][strategy].append(torch.load(os.path.join(path,file_)))
```

```python
base_dir = './visualize'

history={}
subset = ''

# Make save dir
def make_save_dir(base_dir,history):
    dataset = ""
    for data in history_dict.datasets:
        if data in history['config']['train_data_path']:
            dataset = data
            break
    assert dataset != '', "Unknown dataset."

    strategy = history['strategy'] if history['config']['equal_split'] else 'random_'+history['strategy']
    plots_dir = os.path.join(base_dir,dataset,strategy)
    strat_value = str(history['config'][history['strategy']])

    config_str = strat_value+'_LC_'+str(len(history['config']['labeled_classes']))  +'_UC_'+str(len(history['config']['unlabeled_classes']))+\
                '_LS_'+str(history['labeled_set_size'][0])+'_US_'+str(history['unlabeled_set_size'][0])
    plots_dir = os.path.join(plots_dir,config_str)
    os.makedirs(plots_dir,exist_ok=True)

    return plots_dir#, config_str
```

```python
# Function to label the plot in axes coordinates
def prob_label(x, color, label):
    ax = plt.gca()
    
    ax.text(0.05, .3, str(int(label)+1), fontsize='small',#color=color,# fontweight="bold",
        ha="left", va="center", transform=ax.transAxes)
    min_ = '%.2f  '   %(history[subset+'_prob_min'][int(label)])
    max_ = '%.2f    ' %(history[subset+'_prob_max'][int(label)])
    std  = '%.2f   '  %(history[subset+'_prob_std'][int(label)])
    mean = '%.2f'     %(history[subset+'_prob_mean'][int(label)])
    stat = min_+max_+std+mean
    ax.text(1, .3, stat, fontsize='small',#color=color,# fontweight="bold",
        ha="left", va="center", transform=ax.transAxes)

# Function to label the plot in axes coordinates
def sel_prob_label(x, color, label):
    ax = plt.gca()
    
    ax.text(0.05, .3, str(int(label)+1), fontsize='small',#color=color,# fontweight="bold",
        ha="left", va="center", transform=ax.transAxes)
    min_ = '%.2f  '   %(history['selected_prob_min'][int(label)])
    max_ = '%.2f    ' %(history['selected_prob_max'][int(label)])
    std  = '%.2f   '  %(history['selected_prob_std'][int(label)])
    mean = '%.2f'     %(history['selected_prob_mean'][int(label)])
    stat = min_+max_+std+mean
    ax.text(1, .3, stat, fontsize='small',#color=color,# fontweight="bold",
        ha="left", va="center", transform=ax.transAxes)
```

```python
# Function to label the plot in axes coordinates
def class_label(x, color, label):
    ax = plt.gca()
    if subset.lower() == 'selected':
        txt = history['selected_set_size'][int(label)]
        in_do = ' '*(3-len(str(txt)))+'   '+str(history['selected_in_domain'][int(label)]) +' '*(3-len(str(history['selected_in_domain'][int(label)])))
        out_do = '   '+str(history['selected_out_domain'][int(label)])
        txt = str(txt) + in_do + out_do
    elif subset.lower() == 'labeled' or subset.lower() == 'train':
        txt = history['labeled_set_size'][int(label)]
    elif subset.lower() == 'unlabeled' or subset.lower() == 'predicted':
        txt = history['unlabeled_set_size'][int(label)]
    elif subset.lower() == 'pseudo':
        txt = history['selected_set_size'][int(label)]
    else: raise NotImplementedError
    
    ax.text(-0.15, .2, str(int(label)+1), fontsize='small', ha="left", va="center", transform=ax.transAxes)
    ax.text(1.05, .2, str(txt), fontsize='small', ha="left", va="center", transform=ax.transAxes)
```

### 3.3 Metric plots

#### Accuracy Comparison

##### Seaborn

```python
data = {'acc':[], 'name':[], 'idx':[], 'pc_k':[], 'labeled_size':[], 'unlabeled_size':[], 'unlabeled_classes':[]}

# Make data
for history in history_list:
    acc = []
    idx = []
    for val_accs in history['val_acc']:
        acc.append(max(val_accs))
        idx.append(val_accs.index(acc[-1]))
    data['acc'].append(acc)
    data['idx'].append(idx)
    data['pc_k'].append(history['config'].pc_top_k)
    data['labeled_size'].append(history['labeled_set_size'][0])
    data['unlabeled_size'].append(history['unlabeled_set_size'][0])
    data['unlabeled_classes'].append(len(history['config'].unlabeled_classes))
    data['name'].append('k:'+str(history['config'].top_k)+' L:'+str(history['labeled_set_size'][0])+\
        ' UC:'+str(len(history['config'].unlabeled_classes))+' U:'+str(history['unlabeled_set_size'][0]))
```

```python
selections = [
            {'labeled_size':[], 'pc_k':[70], 'unlabeled_classes':[7],  'unlabeled_size':[5250]},
            {'labeled_size':[], 'pc_k':[35], 'unlabeled_classes':[7],  'unlabeled_size':[5250]},
            {'labeled_size':[], 'pc_k':[70], 'unlabeled_classes':[14], 'unlabeled_size':[10500]},
            {'labeled_size':[], 'pc_k':[35], 'unlabeled_classes':[14], 'unlabeled_size':[10500]},
            {'labeled_size':[70,350,1050], 'pc_k':[], 'unlabeled_classes':[7],  'unlabeled_size':[5250]},
            {'labeled_size':[70,350,1050], 'pc_k':[], 'unlabeled_classes':[14], 'unlabeled_size':[10500]},

            # {'labeled_size':[70,350,1050], 'k':[250], 'unlabeled_classes':[], 'unlabeled_size':[5250]},
            # {'labeled_size':[70,350,1050], 'k':[500], 'unlabeled_classes':[], 'unlabeled_size':[10500]},
            ]

max_itr = max([len(accs) for accs in data['acc']])

# plot and save
for selection in selections:
    # populate selection
    order = []
    fixed = []
    variables = []
    for k,v in selection.items():
        if len(v)==0:
            selection[k] = list(set(data[k]))
            variables.append(k)
        elif len(v)>1:variables.append(k)
        else: fixed.append(k)

    # Make Dataframe   
    df = pd.DataFrame()     
    for i,name in enumerate(data['name']):
        if data['pc_k'][i] in selection['pc_k']:
            if data['labeled_size'][i] in selection['labeled_size']:
                if data['unlabeled_classes'][i] in selection['unlabeled_classes']:
                    if data['unlabeled_size'][i] in selection['unlabeled_size']:
                        order.append(''.join([str(data[v][i]) for v in variables]))
                        name = ' ; '.join([v.title()+': '+str(data[v][i]) for v in variables])
                        pad = (max_itr - len(data['acc'][i]))*[np.nan]
                        df[name]=data['acc'][i]+pad

    if df.shape[1]==0:
        print('Skipping:',selection)
        continue
    
    # display(df.head())

    # Set plot size
    plt.figure(figsize=(12,6))
    # Plot the lines
    # dashes=['solid','dashed','solid','dashed','solid','dashed',]
    # style = {key:value for key,value in zip(data[style_col].unique(), dash_list[1:])}

    sns.lineplot(df,markers='o',dashes=False,palette=sns.color_palette("husl",df.shape[1]))
    # Adjust border
    sns.despine(top=True,right=True)
    # Set Labels, legends and ticks

    fixed_txt = (' '*5).join([f.title()+': '+str(selection[f][0]) for f in fixed])
    variable_txt = ' and '.join([v.title() for v in variables])
    plt.suptitle('Accuracy Plot for varing '+variable_txt)
    plt.title(fixed_txt)
    plt.xlabel("Iteration")
    plt.ylabel("Accuracy")
    # Sort Legend
    handles, labels = plt.gca().get_legend_handles_labels() 
    order = [labels.index(lab) for lab in sorted(labels,reverse=True)]
    plt.legend([handles[idx] for idx in order],[labels[idx] for idx in order],facecolor="w",edgecolor='b')
    # Plot / Save
    fixed_txt = '_'.join([f+':'+str(selection[f][0]) for f in fixed])
    variable_txt =  '_'.join(variables)
    # plt.savefig(base_dir+"/varing_["+variable_txt+"]_fixed_["+fixed_txt+"]_acc_plot.png",bbox_inches="tight");
    # plt.close();
    plt.show();
```
